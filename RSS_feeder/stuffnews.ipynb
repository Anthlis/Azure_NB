```python
import feedparser
import bs4
import pprint as pprint
pp = pprint.PrettyPrinter(width=200, compact=False)
```

```python
url = 'https://stuff.co.nz/rss'
feed = feedparser.parse(url)
pp.pprint(feed)
```

```python
feed.get('entries')[0].get('summary')
```

```python
summaries = [item.get('summary') for item in feed.get('entries')]
pp.pprint(summaries)
```
