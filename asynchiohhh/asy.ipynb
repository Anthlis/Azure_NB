```python
import asyncio
import colorama

async def w(n):
    print(colorama.Fore.YELLOW + f" -- sleeping for {n}s!", flush=True)
    await asyncio.sleep(n)
    print(colorama.Fore.BLUE + f" -- done sleeping for {n}s!", flush=True)
    
async def main():
    a1 = w(1)
    a2 = w(2)
    
    await a2
    await a1
    
    print()
    
    t1 = asyncio.create_task(w(1))
    t2 = asyncio.create_task(w(2))
    
    await t2
    await t1
    
asyncio.run(main())    # asynchio.run is a provisional python 3.7+ construct only. Azure NB only 3.6. Sad face.
```
